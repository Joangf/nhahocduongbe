```python
import psycopg2
from sqlalchemy import create_engine
import pandas as pd
import glob
import datetime as dt
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

df = pd.read_csv('20230531_100353.csv')
```

```python
%%capture
df2 = df.iloc[:, :46]
df2.drop(columns='boi_verni', inplace=True)
df2['khoi'] = df['khoi']

# ngay_kham
df2['ngay_kham'][df2['ngay_kham']=='0222-12-02 00:00:00']='2022-12-02 00:00:00'
df2['ngay_kham'][(df2['ngay_kham']=='2012-12-09 00:00:00')|(df2['ngay_kham']=='2202-12-09 00:00:00')|(df2['ngay_kham']=='2202-11-09 00:00:00')]='2022-12-09 00:00:00'
df2['ngay_kham'][df2['ngay_kham']=='0222-12-01 00:00:00']='2022-12-01 00:00:00'
df2['ngay_kham'] = pd.to_datetime(df2['ngay_kham'])

# y_bs_kham
df2['y_bs_kham']='BVRHMTW'

# ten_truong_tieu_hoc
df2['ten_truong_tieu_hoc'][df2['ten_truong_tieu_hoc']=='2022-12-09 00:00:00']='Trung Hiep B'

# ma_so
df2['ma_so']=df2['ma_so'].astype('int')
df2['ma_so'][df2['ma_so']==4888]=488

# nam_sinh
df2['nam_sinh'][df2['nam_sinh'].isnull()]=2015.0
df2['nam_sinh'][df2['nam_sinh']==2914.]=2014.0
df2['nam_sinh']=df2['nam_sinh'].astype('int')

# tuoi
df2['tuoi']=df2['ngay_kham'].dt.year-df2['nam_sinh']

# so_the_bhyt
df2['so_the_bhyt'] = df2['so_the_bhyt'].fillna(-1)
df2['so_the_bhyt']=df2['so_the_bhyt'].astype('int')
df2['so_the_bhyt']=df2['so_the_bhyt'].astype('str')
df2['so_the_bhyt'][df2['so_the_bhyt']=='-1']=np.nan
df2['so_the_bhyt']=df2['so_the_bhyt'].astype('object')

# rangvinhvien_co_bi_sau
df2['rangvinhvien_co_bi_sau'][df2['rangvinhvien_co_bi_sau']=='0.0']='khong'

# r16
df2['r16'][df2['r16']=='0.0']='khong'
df2['r16'][df2['r16']=='3.0']='co'
df2['r16'][df2['r16']=='8.0']='khong'

# r26
df2['r26'][df2['r26']=='0.0']='khong'
df2['r26'][df2['r26']=='8.0']='khong'

# r36
df2['r36'][df2['r36']=='0.0']='khong'

# r46
df2['r46'][df2['r46']=='8.0']='khong'

# rvv_khac
df2['rvv_khac'][df2['rvv_khac']=='0.0']='khong'

# soluongrangsua_da_tram
df2['soluongrangsua_da_tram'][df2['soluongrangsua_da_tram']==52.0]=1.0

# r1716_ngoai
df2['r1716_ngoai'][df2['r1716_ngoai']=='khong']=0
df2['r1716_ngoai'][df2['r1716_ngoai']=='mot phan ba']=1
df2['r1716_ngoai'][df2['r1716_ngoai']=='hai phan ba']=2
df2['r1716_ngoai'][df2['r1716_ngoai']=='hon hai phan ba']=3
df2['r1716_ngoai'][df2['r1716_ngoai']=='khong co rang']=4

# r11_ngoai
df2['r11_ngoai'][df2['r11_ngoai']=='khong']=0
df2['r11_ngoai'][df2['r11_ngoai']=='mot phan ba']=1
df2['r11_ngoai'][df2['r11_ngoai']=='hai phan ba']=2
df2['r11_ngoai'][df2['r11_ngoai']=='hon hai phan ba']=3
df2['r11_ngoai'][df2['r11_ngoai']=='khong co rang']=4

# r2627_ngoai
df2['r2627_ngoai'][df2['r2627_ngoai']=='khong']=0
df2['r2627_ngoai'][df2['r2627_ngoai']=='mot phan ba']=1
df2['r2627_ngoai'][df2['r2627_ngoai']=='hai phan ba']=2
df2['r2627_ngoai'][df2['r2627_ngoai']=='hon hai phan ba']=3
df2['r2627_ngoai'][df2['r2627_ngoai']=='khong co rang']=4

# r3637_trong
df2['r3637_trong'][df2['r3637_trong']=='khong']=0
df2['r3637_trong'][df2['r3637_trong']=='mot phan ba']=1
df2['r3637_trong'][df2['r3637_trong']=='hai phan ba']=2
df2['r3637_trong'][df2['r3637_trong']=='hon hai phan ba']=3
df2['r3637_trong'][df2['r3637_trong']=='khong co rang']=4

# r31_ngoai
df2['r31_ngoai'][df2['r31_ngoai']=='khong']=0
df2['r31_ngoai'][df2['r31_ngoai']=='mot phan ba']=1
df2['r31_ngoai'][df2['r31_ngoai']=='hai phan ba']=2
df2['r31_ngoai'][df2['r31_ngoai']=='hon hai phan ba']=3
df2['r31_ngoai'][df2['r31_ngoai']=='khong co rang']=4

# r4746_trong
df2['r4746_trong'][df2['r4746_trong']=='khong']=0
df2['r4746_trong'][df2['r4746_trong']=='mot phan ba']=1
df2['r4746_trong'][df2['r4746_trong']=='hai phan ba']=2
df2['r4746_trong'][df2['r4746_trong']=='hon hai phan ba']=3
df2['r4746_trong'][df2['r4746_trong']=='khong co rang']=4

# khoi
df2['khoi'][df2['khoi']=='khối 1']=1
df2['khoi'][df2['khoi']=='khối 2']=2
df2['khoi'][df2['khoi']=='khối 3']=3
df2['khoi'][df2['khoi']=='khối 4']=4
df2['khoi'] = df2['khoi'].astype('int')

```
